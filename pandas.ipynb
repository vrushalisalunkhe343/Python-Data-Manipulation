# joinnig and merging

```python
Amazon=pd.DataFrame({'Product_Name':['shooes','lipstick','Top','saree'],
                    'deliver_Location':['pune','mumbai','nashik','banglore'],
                    'product_price':[1500,3000,6789,1200]
})
Amazon
```

```python
Filfkart=pd.DataFrame({'Product_Name':['shirt','pants','hair_mask','face_mask'],
                      'deliver_Location':['dhule','jalgaon','indor','chalisgaon'],
                      'product_price':[1230,4500,5698,8765]})
Filfkart
```

```python
pd.concat([Amazon,Filfkart],ignore_index=True)
```

```python
pd.concat([Amazon,Filfkart],axis=1)
```
